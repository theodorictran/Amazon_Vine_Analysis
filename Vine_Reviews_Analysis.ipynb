```python
import numpy as np
import pandas as pd
```

```python
# reading vine_table csv
df = pd.read_csv("vine_table.csv", delimiter=",")
```

```python
# previewing dataframe
df.head()
```

```python
# creating a dataframe based on total_votes >= 20.0
total_votes_df = df[df.total_votes >= 20.0]
total_votes_df.head()
```

```python
# filtered and created a dataframe where helpful votes divided by total votes is >= 50%
filtered_votes_df = total_votes_df[total_votes_df.helpful_votes/total_votes_df.total_votes >= .5]
filtered_votes_df.sample(5)
```

```python
# created a dataframe based on filtered_votes_df to return reviewers part of the Vine program
yes_vine_df = filtered_votes_df[filtered_votes_df.vine == "Y"]
yes_vine_df.sample(5)
```

```python
# created a dataframe based on filtered_votes_df to return reviewers NOT part of the Vine program
not_vine_df = filtered_votes_df[filtered_votes_df.vine == "N"]
not_vine_df.sample(5)
```

```python
# Determine the total number of reviews, the number of 5-star reviews,
# and the percentage of 5-star reviews for the two types of review (paid vs unpaid).

total_paid_reviews = yes_vine_df.shape[0]
total_unpaid_reviews = not_vine_df.shape[0]
num_five_star_reviews_paid = yes_vine_df[yes_vine_df.star_rating == 5.0].shape[0]
num_five_star_reviews_unpaid = not_vine_df[not_vine_df.star_rating == 5.0].shape[0]
percentage_five_star_reviews_paid = num_five_star_reviews_paid/total_paid_reviews
percentage_five_star_reviews_unpaid = num_five_star_reviews_unpaid/total_unpaid_reviews


print("----------------------------------")
print("Paid Reviews")
print(f"Total Number of Reviews: {total_paid_reviews}")
print(f"Number of 5-Star Reviews: {num_five_star_reviews_paid}")
print(f"Percentage of 5-Star Reviews: {percentage_five_star_reviews_paid:.2f}")
print("----------------------------------")
print("Unpaid Reviews")
print(f"Total Number of Reviews: {total_unpaid_reviews}")
print(f"Number of 5-Star Reviews: {num_five_star_reviews_unpaid}")
print(f"Percentage of 5-Star Reviews: {percentage_five_star_reviews_unpaid:.2f}")
print("----------------------------------")
```
